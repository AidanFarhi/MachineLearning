```python
import numpy as np
import matplotlib.pyplot as plt
```

## Standard Deviation

The square root of the average of the squared differences from the mean.

## Variance
The average of the squared differences from the mean.

```python
data = np.random.normal(50, 1000, 10_000)
```

```python
plt.hist(data, 50)
plt.show()
```

```python
data.std()
```

```python
data.var()
```
