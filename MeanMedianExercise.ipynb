```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
transactions = np.random.normal(1_000, 100, 10_000)
```

```python
plt.hist(transactions, 50)
plt.show()
```

```python
print('mean:', np.mean(transactions))
print('median:', np.median(transactions))
print('mode:', stats.mode(transactions, keepdims=True))
```
